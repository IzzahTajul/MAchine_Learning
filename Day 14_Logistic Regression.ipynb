```python
# Sigmoid Fx aka logistic fx
# give S shape curve that can take real valued number and map it into value
# between 0 and 1
# more positivity, becomes 1 or YES
# curve is more 0.5 --> YES, less than 0.5 -->NO
# curve is more 0.75 --> patient high chance got cancer
```

```python
# Supervised ML
# Classification --> Logistic Regression

# Sigmoid Function

import numpy as np
from sklearn.linear_model import LogisticRegression

# Step 2: Create the dataset
X = np.array([[1],[2],[3],[4],[5],[6]])   #hours_studied
y = np.array([0,0,0,1,1,1])               #passed or not
y
```

```python
# Step 3: Create and train the model
model = LogisticRegression()
model.fit(X,y)
```

```python
# predict_proba() stands for predict probabilities
y_prob =model.predict_proba(X)[:,1]    #[:,1] all rowa but column 1
y_prob
```

```python
# Get the slope and intercept
a = model.coef_ [0][0]    #slope       coef_ is a 2D array in scikit-learn
b = model.intercept_[0]   #intercept

print(f'a (slope):{a}')
print(f'b (intercept):{b}')

# a=1.12means model thinks 'hours_studied' strongly
# affects the chance of passing
# if we had a = 0.2, effect would be much weaker

# b = -3.9 means a student who studies very few hours has a very low chance of passing.
```

```python
# Compute the score --> if student studying for 3.5  hours
a = 1.1207
b = -3.9223
x = 3.5

score = a*x + b
print(score)
```

```python
# Step 2 : Turn score into probability (Using Sigmoid)

import math
res = math.e** 0.0015
print(res)
```

```python
res = 1/(1+math.e** 0.0015)
print(res)

# 0.49 --> 49% probability, there is 49% chance student will pass
# 0.49 <0.5 --> predict 'Fail'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.info()
```

```python
# Missing value calculation
data.isnull().sum()
```

```python
# Duplicates check up
data.duplicated().sum()
```

```python
# select independent and dependent variable
data.columns
```

```python
x = data.drop('Outcome', axis=1)  #axis = 1 means column
y = data['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# predict_proba() stands for predict probabilities
# predict_proba() returns 2 column for negative and positive class
# get the predicted probabilities of the positive label only
y_prob =model.predict_proba(x_test)[:,1]
y_prob
```

```python
# get the predicted values
y_pred = model.predict(x_test)
y_pred
```

```python
# Get predictions
y_prob = model.predict_proba(x_test)[:, 1]
y_pred = model.predict(x_test)

# Create a table
results = pd.DataFrame({
    'Predicted_Probability': y_prob,
    'Predicted_Class': y_pred
})

# Show first few rows
print(results.head(10))
```

```python
evl1 = pd.DataFrame({
    'Probabilities': y_prob,
    'Predicted_value': y_pred
})
evl1.head(50)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
cm

#                          Actual_DB               NON_DB
#    Predicted_DB           78  TP                  21 FP
#    Predicted_Non-DB       18  FP                  37 TP

#                             Total No. OB = 154    115 Correct  39 wrong
```

```python
y_pred.shape
```

```python
sns.heatmap(cm,annot=True, fmt='d',cmap='Blues')  # fmt -- d means display decimal/int value
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Results from confusion matrix
# Accuracy: How oftern model was correct


# Precision: When model says its spam, how often it actually spam
# when model say someone has DB, can i trust it?
#imagine 10 pple have DB, but in reality, only 8 are DB
# Precision = TP/(TP+FP)
#10/8 -> 0.8, means when model says YES ---> it is correct 80% of the time

# Recall (sensitivity): of all actual spam, how many did model correctly identify
# Did my model miss diabetic pple?
# Recall = TP+FN/TP
#8/12 = 0.67
# model found 67& of all actual DB
# Ability to find all positives

# F1 Score
```
